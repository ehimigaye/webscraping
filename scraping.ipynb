```python
#Exercice :
import requests
import string
from bs4 import BeautifulSoup
#question 1 :
wiki = requests.get("https://fr.wikipedia.org/wiki/S%C3%A9n%C3%A9gal")
n = BeautifulSoup(wiki.content,'html.parser')
# contenu du code html à analyser :
print(n.prettify())
```

```python
#Question 2 :
#extraire le titre de l' article :
print(n.title)
print(n.title.text)

```

```python
#Question 4 :
#extraire les liens de la pages :
for link in n.find_all('a'):
    print(link.get('href'))
```

```python
#Question 3:
#extraire les titres et les paragraphes
doc= n('div',class_= "mw-body-content mw-content-ltr")
for i in doc:
    for h1 in i.find_all('h1'):
        print("Nos H1 : ",h1.text)
        print("-"*100)
    for p in i.find_all('p'):
        print(p.text)
        
   


  
```

```python
#Question 5:
#creons la fonction et ensuite nous la testerons 
import requests
import string
from bs4 import BeautifulSoup
recherche = input()
u = string.capwords(recherche)
l = u.split()
r = "_".join(l)
url = "https://fr.wikipedia.org/wiki/"+r
print(url)
def exercice(url):
    lien = requests.get(url)
    N = BeautifulSoup(lien.content,'html.parser')
    #extraction et analyse du code html de la page
    print(N.prettify())
    #extraction du titre
    print(N.title.text)
    #extraire les liens de la page
    for link in n.find_all('a'):
    print(link.get('href'))
    doc= N('div',class_= 'mw-content-container')
    for i in doc:
        for h1 in i.find_all('h1'):
            print("Nos H1 : ",h1.text)
            print("-"*100)
        for p in i.find_all('p'):
            print(p.text)
        
    
exercice(url)
#testons la fonction avec le maroc
```
